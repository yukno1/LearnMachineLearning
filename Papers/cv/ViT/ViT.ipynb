## 公式
输入patch（(P\*P)\*C），全连接得到（P\*P\*D）

$$
\begin{aligned}
{\bf z}_0 &=[{\bf x}_{{\rm class}};{\bf x}_{p}^1\bf{E};\dots;\bf{x}_{p}^N{\bf E}]+{\bf E}_{pos},&&\bf{E}\in {\mathbb R}^{(P^{2}\cdot C)\times D},\bf{E}_{pos}\in {\mathbb R}^{(N+1)\times D} \\
\mathbf{z}'_{\mathscr{l}}&=\mathrm{MSA}(\mathrm{LN}(\mathbf{z}_{\mathscr{l}-1}))+\mathbf{z}_{\mathscr{l}-1},&&\mathscr{l}=1,\dots,L \\
\bf{z}_{\scr{l}}&=\rm{MLP}(\rm{LN}(\bf{z}_{\scr{l}}))+\bf{z}_{\scr{l}},&&{\scr l}=1,\dots,L \\
\bf{y}&=\rm{LN}(\bf{z}_{L}^0)
\end{aligned}
$$

${\rm E}$ 是$196\times 256$全连接矩阵，每个序列196个token，每个token是256维向量

multihead self attention
